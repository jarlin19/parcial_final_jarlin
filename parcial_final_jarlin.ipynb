```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Cargar el dataset
df = pd.read_csv("dataset.csv")

# Mostrar las primeras filas
df.head()
```

```python
# Información general
df.info()

# Número de valores nulos por columna
print("\nValores nulos por columna:\n")
print(df.isnull().sum())

# Estadísticas descriptivas numéricas
df.describe()

```

```python
# Género
sns.countplot(x="genero", data=df)
plt.title("Distribución por género")
plt.show()

# País
plt.figure(figsize=(12,5))
sns.countplot(x="pais", data=df, order=df["pais"].value_counts().index)
plt.title("Distribución por país")
plt.xticks(rotation=45)
plt.show()

# Estado de suscripción
sns.countplot(x="estado_suscripcion", data=df)
plt.title("Estado de suscripción")
plt.show()
```

```python
# Histograma de edad
sns.histplot(df["edad"], bins=30, kde=True)
plt.title("Distribución de edad")
plt.show()

# Ingreso anual
sns.histplot(df["ingreso_anual"], bins=30, kde=True)
plt.title("Distribución de ingreso anual")
plt.show()

# Número de compras
sns.histplot(df["numero_compras"], bins=30, kde=True)
plt.title("Distribución del número de compras")
plt.show()

# Valoración del cliente
sns.countplot(x="valoracion_cliente", data=df)
plt.title("Valoración de los clientes")
plt.show()
```

```python
# Matriz de correlación
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Matriz de correlación")
plt.show()
```

```python
# Relación entre edad e ingreso
sns.scatterplot(data=df, x="edad", y="ingreso_anual", hue="genero")
plt.title("Edad vs Ingreso Anual")
plt.show()

# Relación entre número de compras y valor promedio
sns.scatterplot(data=df, x="numero_compras", y="valor_promedio_compra", hue="estado_suscripcion")
plt.title("Compras vs Valor promedio de compra")
plt.show()
```
